5.2 Выберите ноутбук клиенту:  

Клиент хочет подобрать ноутбук от MSI, с видеокартой Nvidia GeForce GTX 1050 Ti и главное не с диагональю 15.6.  
В какой ценовой категории вышли подобные ноутбуки?  

```python
msi_df = df[(df['Company'] == 'MSI') & (df['Gpu'] == 'Nvidia GeForce GTX 1050 Ti') & (df['Inches'] != 15.6)]
msi_df
```

5.3 Что дешевле?  

В среднем дешевле ноутбуки с CPU Intel Core i7 7700HQ 2.8GHz или с Intel Core i7 7600U 2.8GHz?

```python
ic_1 = df[df['Cpu'] == 'Intel Core i7 7700HQ 2.8GHz']['Price_euros'].mean().round(2)
ic_1
```

```python
ic_2 = df[df['Cpu'] == 'Intel Core i7 7600U 2.8GHz']['Price_euros'].mean().round(2)
ic_2
```

```python
if ic_1 < ic_2:
    print('Дешевле ноутбук с CPU Intel Core i7 7700HQ 2.8GHz')
else:
    print('Дешевле ноутбук с CPU Intel Core i7 7600U 2.8GHz')
```

```python
# Найдем вес самого легкого ноутбука
df['Weight'].min()
```
