Процессоры от Samsung не изучайте  

3.1 Постройте график в абсолютных величинах  
3.2 Постройте график в относительных величинах  


```python
df.sample()
```

```python
df2 = df[df['Cpu_Company'] != 'Samsung']
```

```python
ct = pd.crosstab(index=df2['Company'], columns=df2['Cpu_Company']).reset_index()
# Метод .reset_index() - сбрасывает индекс
ct
```

## 3.1 Постройте график в абсолютных величинах  


```python
plt.figure(figsize=(20,8))
n_ticks = ct.index
bar_w = 0.4 #ширина столбца
offset = 0.2 #смещение относительно шкалы
plt.bar(n_ticks-offset, ct['Intel'], width=bar_w)
plt.bar(n_ticks+offset, ct['AMD'], width=bar_w)
plt.title('Ноутбуки и процессоры')
plt.legend(['Intel', 'AMD'])
plt.xticks(n_ticks, ct['Company']) 

;
```

## 3.2 Постройте график в относительных величинах  

```python
ct = pd.crosstab(index=df2['Company'], columns=df2['Cpu_Company'], normalize='index').reset_index().round(2)
# Метод .reset_index() - сбрасывает индекс
ct
```

```python
plt.figure(figsize=(20,4))
plt.bar(ct['Company'], ct['Intel'])
plt.bar(ct['Company'],  ct['AMD'], botom=ct['Intel'])
plt.title('Ноутбуки и процессоры относительно')
plt.legend(['Intel', 'AMD'])
plt.yticks(np.arange(0, 1.1, 0.1), np.arange(100, 110, 10)) #перевели в проценты

;
```
